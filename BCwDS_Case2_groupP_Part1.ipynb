Due to the lack of information about the macroeconomic indicatores used, our analysis was based on the assumption that the following definitions are correct.


- The **Production index** measures changes in value added at factor cost of industry and construction over a month. It does this by measuring changes in the volume of output and activity monthly.
- The **Shipment index** is a measure of the volume of goods shipped by manufacturers monthly. It is an indicator of economic trends, since it reflects production, supply chain dynamics and overall industrial performance. 
- The **World prices of materials** is monthly average global market prices for key industrial inputs, such as energy, heavy materials, etc. These are likely to fluctuate with changes in supply/demand dynamics and economic conditions. 
- The **United States: EUR in LCU** is the monthly exchange rate of the Euro, expressed in local currency units from the United States (US dollar), representing the number of US dollars per euro. It indicates currency fluctuation, which often impacts trade.
- The **Producer prices** measures the monthly change in the trading price of products and related services, from the seller's perspective. It does not only serve as an early indicator of inflationary pressures in the economy before it reaches the consumer, but it can also record the evolution of prices over longer time periods.

```python
def save_correlation_matrices(product_dfs, threshold=0.90, cmap='YlGnBu', save_path='corr_matrices'):
    os.makedirs(save_path, exist_ok=True)  # Create directory for images

    image_paths = []
    for key in product_dfs.keys():
        data = product_dfs[key]
        corr = data.corr().round(2)

        # Generate annotation mask
        mask_annot = np.absolute(corr.values) >= threshold
        annot = np.where(mask_annot, corr.values, np.full(corr.shape, ''))

        # Plot correlation matrix
        fig, ax = plt.subplots(figsize=(5, 5))  # Adjust size as needed
        sns.heatmap(data=corr, annot=annot, cmap=cmap, fmt='s', vmin=-1, vmax=1, center=0, 
                    square=True, linewidths=0.5, annot_kws={'size': 10}, ax=ax)
        
        ax.set_title(f'Correlation Matrix - {key}', fontsize=12)
        plt.xticks(rotation=45, ha='right', fontsize=8)
        plt.yticks(fontsize=8)

        # Save image
        img_path = os.path.join(save_path, f"{key}.png")
        plt.savefig(img_path, bbox_inches='tight')
        plt.close(fig)  # Close figure to free memory

        image_paths.append(img_path)
    
    return image_paths

def display_images_side_by_side(image_paths, tablespacing=5, width="auto", height="500px"):
    """Displays images in a scrollable horizontal div"""
    html = """
    <div style='overflow-x: auto; white-space: nowrap; padding-bottom: 10px; border-bottom: 2px solid #ddd;'>
    """
    for img_path in image_paths:
        html += f"""
        <div style='display:inline-block; margin-right:{tablespacing * 5}px; vertical-align:top;'>
            <img src='{img_path}' style='width:{width}; height:{height}; border:1px solid #ccc; padding:5px;'>
        </div>
        """
    html += "</div>"
    display(HTML(html))
```

```python
#plot the sales data over a period of time
def plot_sales(dataframe, x_column, y_column, title, x_label, y_label):
    fig = px.line(dataframe, x=x_column, y=y_column,title=title, 
                  labels={x_column: x_label, y_column: y_label})

    fig.show()
```

```python
#plot the sales data by product
def plot_sales_by_product(dataframe, x_column, y_column, title, x_label, y_label):
    fig = px.bar(dataframe, x=x_column, y=y_column, title=title, 
                labels={x_column: x_label, y_column: y_label},
                text=y_column)
    
    fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
    fig.update_layout(xaxis=dict(type='category', tickmode='linear'))
    fig.show()
```

```python
#plot the sales data over a period of time by product
def plot_sales_by_product_month(dataframe, x_column, y_column, category_column, title, x_label, y_label):
    
    fig = px.line(dataframe, 
                  x=x_column, 
                  y=y_column, 
                  color=category_column,  # Different lines for each product
                  title=title, 
                  labels={x_column: x_label, y_column: y_label, category_column: "Product"})
    
    fig.update_layout(xaxis=dict(tickangle=-45))  # Rotate x-axis labels for better readability
    fig.show()
```

```python
def plot_sales_trends(data, trend_type='linear'):

    data['month_year'] = pd.to_datetime(data['month_year'])
    unique_products = data['Mapped_GCK'].unique()
    
    fig = make_subplots(
        rows=1, cols=len(unique_products),
        subplot_titles=[f'Product {p}' for p in unique_products],
        horizontal_spacing=0.02
    )
    
    for i, product in enumerate(unique_products):
        product_data = data[data['Mapped_GCK'] == product].copy()
        product_data = product_data.sort_values(by='month_year')
        
        if trend_type == 'linear':
            product_data['month_numeric'] = product_data['month_year'].map(pd.Timestamp.toordinal)
            slope, intercept, _, _, _ = linregress(product_data['month_numeric'], product_data['Sales_EUR'])
            product_data['Trend'] = intercept + slope * product_data['month_numeric']
        else:  # Rolling mean trend
            product_data['Trend'] = product_data['Sales_EUR'].rolling(window=3, min_periods=1).mean()
        
        fig.add_trace(
            go.Scatter(x=product_data['month_year'], y=product_data['Sales_EUR'],
                       mode='lines+markers', name=f'Product {product} Sales',
                       line=dict(color='blue')),
            row=1, col=i+1)
        
        fig.add_trace(
            go.Scatter(x=product_data['month_year'], y=product_data['Trend'],
                       mode='lines', name=f'Product {product} Trend',
                       line=dict(color='red', dash='dot')),
            row=1, col=i+1)
    
    fig.update_layout(
        title="Positive Sales Trends by Product (Monthly)",
        xaxis_title="Year-Month",
        yaxis_title="Total Positive Sales (EUR)",
        showlegend=False,
        height=500, width=500 * len(unique_products),
        xaxis=dict(tickangle=-45)
    )
    
    fig.show()
```


‚û° What is the "Stationarity"? The distribution of the random variables in the time-series do not change over time. Meaning, the mean and variance do not change if you go further in time or at any time point.


‚û° Two Types of Stationarity:
- 1. **Strong Sense Stationarity (SSS):** *"The entire probability distribution does not change over time."* SSS is rarely useful in real-world applications because it's an extremely strict assumption.

- 2. **Weak Sense Stationarity (WSS):** *"Only the first-order (mean) and second-order (autocovariance) statistics remain constant over time."* This is a **relaxed** version of stationarity and much more practical. Instead of requiring the full distribution to remain unchanged, WSS only ensures that:

    - The **mean** of the series is constant over time.
    - The **autocovariance** (dependence between different time points) depends only on the time difference (lag) and not on the actual time.

WSS is what we usually test for when analyzing time-series data in statistics and machine learning.

###### <span style="background-color:#000027; padding:5px; border-radius:5px;font-size:13pt">**2.6.1. Dickey-Fuller Test**</span> <a id='Dickey-Fuller'></a> 

- Null Hypothesis: Time series is non-stationary
- Alternative Hypothesis: Time series is stationary.

A p-value below 0.05 indicates that we reject the null hypothesis, suggesting the time series is stationary.

```python
def adf_test(series):
    """Perform Augmented Dickey-Fuller (ADF) test for stationarity."""
    result = adfuller(series.dropna())  # Drop NaNs before test
    print("Dickey-Fuller Test Results:")
    df_results = pd.Series(result[:4], index=["Test Statistic", "p-value", "# Lags Used", "# Observations Used"])
    for key, value in result[4].items():
        df_results[f"Critical Value ({key})"] = value
    print(df_results)
    
    # Interpretation
    if result[1] <= 0.05:
        print("‚úÖ The series is stationary (p-value < 0.05).")
    else:
        print("‚ùå The series is NOT stationary (p-value > 0.05). Consider transforming it.")
```

```python
for col in df_sales.columns[:15]:
    print(f"{col}")
    adf_test(df_sales[col]) 
```

###### <span style="background-color:#000027; padding:5px; border-radius:5px;font-size:13pt">**2.6.2. Autocorrelation Function (ACF)**</span> <a id='Autocorrelation'></a> 

- The Autocorrelation Function (ACF) measures how correlated a time series is with itself over different time lags.
- This helps determine the q parameter in Moving Average (MA) models.

- If ACF values are near zero ‚Üí The time series has no correlation (random).
- If ACF values drop slowly ‚Üí There‚Äôs some correlation over time.
- If some lags have high ACF outside the confidence interval (blue region) ‚Üí Suggests a pattern at that lag.

`Since our data is random noise, we expect the ACF to be close to zero for all lags except lag = 0 (which is always 1 because a series is perfectly correlated with itself).`

```python
def plot_correlation_for_products(data, lags=30, plot_type='acf'):
    """
    Plots ACF or PACF with stem-like bars and constant confidence intervals
    
    Parameters:
    -----------
    lags : int
        Number of lags to compute
    plot_type : str ('acf' or 'pacf')
        Type of correlation plot to display
    """
    # Validate plot_type
    plot_type = plot_type.lower()
    if plot_type not in ['acf', 'pacf']:
        raise ValueError("plot_type must be either 'acf' or 'pacf'")
    
    fig = make_subplots(
        rows=1,
        cols=len(data.columns[:15]),
        subplot_titles=[f"{plot_type.upper()}: {product}" for product in data.columns[:15]],
        horizontal_spacing=0.01)

    for i, product in enumerate(data.columns[:15]):
        clean_data = data[product].fillna(0)
        
        if plot_type == 'acf':
            corr_values = acf(clean_data, nlags=lags, fft=True)
            # Standard error for ACF (Bartlett's formula)
            se = 2.24 / np.sqrt(len(clean_data)) # A 97.5% confidence interval means that (1 - 0.975) / 2 = 0.0125 (or 1.25%) of the probability is in each tail. The corresponding Z-score for 97.5% confidence is 2.24 (approximately).
        else:
            corr_values = pacf(clean_data, nlags=lags)
            # Standard error for PACF (1/sqrt(n))
            se = 2.24 / np.sqrt(len(clean_data))
            
        lags_range = list(range(lags+1))
        
        # Create stem-like effect (line + marker at top)
        fig.add_trace(
            go.Scatter(
                x=lags_range,y=corr_values,mode='lines+markers',line=dict(color='blue', width=1),
                marker=dict(color='blue', size=5),name=product,
                hovertemplate=f"Lag %{{x}}<br>{plot_type.upper()}: %{{y:.2f}}<extra></extra>"),
            row=1, col=i+1)
        
        # Add thin vertical lines from 0 to correlation value (stem effect)
        for lag, val in zip(lags_range, corr_values):
            fig.add_trace(
                go.Scatter(
                    x=[lag, lag],y=[0, val],mode='lines',line=dict(color='blue', width=1),
                    showlegend=False,hoverinfo='skip'),row=1, col=i+1)
        
        # Add constant confidence intervals
        fig.add_hline(y=se, line=dict(color='red', width=1, dash='dash'), 
                     row=1, col=i+1, opacity=0.7)
        fig.add_hline(y=-se, line=dict(color='red', width=1, dash='dash'), 
                     row=1, col=i+1, opacity=0.7)
        
        # Shaded confidence interval area
        fig.add_hrect(y0=-se, y1=se, 
                     line_width=0, fillcolor='rgba(100,100,100,0.2)', 
                     row=1, col=i+1)

    fig.update_layout(
        title_text=f"{plot_type.upper()} by Product (97.5% Confidence)",
        height=500,
        width=500*len(data.columns[:15]),
        showlegend=False,
        margin=dict(l=20, r=20))
    
    # Uniform axes settings
    fig.update_yaxes(range=[-1.1, 1.1], title_text=plot_type.upper())
    fig.update_xaxes(title_text="Lag")
    
    fig.show()
```

```python
# Example usage
plot_correlation_for_products(df_sales, plot_type='acf')
```

- Most points fall within the blue confidence interval. No clear pattern or strong lags, this means the series does not have strong dependencies on past values.
- The blue shaded area represents the 95% confidence interval (assuming white noise). If autocorrelation values stay inside this region, they are not statistically significant. If all lags (except lag 0) are close to zero and within the confidence interval, the series could be purely random (white noise).
- Since no strong lags appear, an MA model may not be useful.
- Typically, we select ùëû where the autocorrelation drops off sharply (but here, no clear drop exists).
- The lack of strong lags suggests that the data might already be **stationary**.

###### <span style="background-color:#000027; padding:5px; border-radius:5px;font-size:13pt">**2.6.3. Partial Autocorrelation Function (PACF)**</span> <a id='PACF'></a> 

- PACF at Lag 0 = 1. This is always 1 because a time series is perfectly correlated with itself.
- The first significant spike in PACF suggests the order (p) of the AR model. If PACF cuts off after lag k, then p = k. The rest of the lags should ideally fall within the confidence interval (blue shaded area).

Comparison with ACF:

    - ACF shows how a value is correlated with past values directly and indirectly.
    - PACF shows how a value is correlated only with its direct past, ignoring indirect effects.

- If all PACF values are within the blue region, then there is no AR structure in the data, and an AR model may not be needed.
- Slow decay in ACF? ‚Üí Data is non-stationary

```python
def plot_correlation_matrix(data, columns, title='Correlation Matrix', 
                          threshold=0.8, figsize=(15, 12), cmap='YlGnBu'):
    """
    Plots a styled correlation matrix for specified columns.
    
    Parameters:
    -----------
    columns : list
        List of column names to include in correlation
    title : str
        Title for the plot
    threshold : float
        Only show correlation values above this absolute value
    figsize : tuple
        Figure size (width, height)
    font_scale : float
        Scaling factor for fonts
    shorten_names : bool
        Whether to remove 'MAB_ELE_' prefix from column names
    """
    # Subset and calculate correlations
    corr = data[columns].corr().round(2)
    labels = columns
    
    # Set up the matplotlib figure
    plt.figure(figsize=figsize)
    
    # Generate annotation mask
    mask_annot = np.absolute(corr.values) >= threshold
    annot = np.where(mask_annot, corr.values, np.full(corr.shape, ''))
    
    # Draw the heatmap
    sns.heatmap(data=corr,annot=annot,cmap=cmap,fmt='s',vmin=-1,
        vmax=1,center=0,square=True,linewidths=0,annot_kws={'size': 15},
        xticklabels=labels,yticklabels=labels)
    
    plt.xticks(fontsize=15)  
    plt.yticks(fontsize=15)  
    
    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    
    # Adjust title
    plt.title(title, fontsize=18, pad=20)
    plt.tight_layout()
    plt.show()
```

##### <span style="background-color:#000027; padding:5px; border-radius:5px;">**1.1. Feature Engineering**</span> <a id='feature-engineering-dp'></a>

> Percentage Change

Now we want to calculate the percentage change for each column and create new columns with that percentage:

```python
# Replace all 0 values in df with 1
df_sales_preparation = df_sales_preparation.replace(0, 1) # This ensures that there are no 0s in the DataFrame, which helps prevent division 
                                  # errors when calculating percentage changes.

# Create a list of the column names of df_sales_preparation
nominal_prod_list = list(df_sales_preparation.columns)

# Initialize an empty list to store the names of the columns containing percent changes
percent_change_prod_list = []

# For each column in the DataFrame 'df', calculate the percent change and add a new column with the suffix '%Change'
for col in df_sales_preparation.columns:
    df_sales_preparation[col + '%Change'] = df_sales_preparation[col].pct_change()
    
    # Append the name of the new column to the percent_change_prod_list
    percent_change_prod_list.append(col + '%Change')

# Replace all 1 values in columns specified in 'nominal_prod_list' with 0
df_sales_preparation[nominal_prod_list] = df_sales_preparation[nominal_prod_list].replace(1, 0)
```

###### <span style="background-color:#000027; padding:5px; border-radius:5px;font-size:13pt">**1.4.2. Autocorrelation Function (ACF)**</span> <a id='Autocorrelation'></a> 

- The Autocorrelation Function (ACF) measures how correlated a time series is with itself over different time lags.
- This helps determine the q parameter in Moving Average (MA) models.

- If ACF values are near zero ‚Üí The time series has no correlation (random).
- If ACF values drop slowly ‚Üí There‚Äôs some correlation over time.
- If some lags have high ACF outside the confidence interval (blue region) ‚Üí Suggests a pattern at that lag.

`Since our data is random noise, we expect the ACF to be close to zero for all lags except lag = 0 (which is always 1 because a series is perfectly correlated with itself).`

```python
plot_correlation_for_products(df_sales_cpi, plot_type='acf')
```

###### <span style="background-color:#000027; padding:5px; border-radius:5px;font-size:13pt">**1.4.3. Partial Autocorrelation Function (PACF)**</span> <a id='PACF'></a> 

- PACF at Lag 0 = 1. This is always 1 because a time series is perfectly correlated with itself.
- The first significant spike in PACF suggests the order (p) of the AR model. If PACF cuts off after lag k, then p = k. The rest of the lags should ideally fall within the confidence interval (blue shaded area).

Comparison with ACF:

    - ACF shows how a value is correlated with past values directly and indirectly.
    - PACF shows how a value is correlated only with its direct past, ignoring indirect effects.

- If all PACF values are within the blue region, then there is no AR structure in the data, and an AR model may not be needed.
- Slow decay in ACF? ‚Üí Data is non-stationary

```python
plot_correlation_for_products(df_sales_cpi, lags=21, plot_type='pacf')
```

#### <span style="background-color:#000027; padding:5px; border-radius:5px;">**4. Feature Selection**</span> <a id='feature-selection'></a>  

- Pearson Correlation: Measures linear relationships between features and target ‚Üí Identifies features with strong linear predictive power.
- Spearman Correlation: Measures monotonic (non-linear but ordered) relationships ‚Üí Captures non-linear but consistently increasing/decreasing patterns.
-  Tree-Based Importance (Random Forest & XGBoost): 
    - Random Forest Importance: Uses Gini impurity reduction to rank features, is robust to outliers and handles non-linearities.
    - XGBoost Importance: Uses gain (improvement in accuracy) from splits, better for gradient boosting‚Äôs focus on error correction.
- SHAP Values (XGBoost-Based): Quantifies contribution of each feature to predictions (direction and magnitude). The result, mean absolute SHAP value per feature ‚Üí higher = more impactful.
- Single-Feature Predictive Power (RMSE/R¬≤): using a simple linear regression for each feature individually ‚Üí Identifies features that are independently predictive (no multicollinearity considered); RMSE (lower = better) and R¬≤ (higher = better),

```python

def feature_selection(df, target_col, top_n=5):
    """
    Performs feature selection by calculating various importance metrics and
    marking top_n features per method as 'keep'.
    """
    # Prepare X and y
    X = df.drop(columns=[target_col])
    y = df[target_col]

    # Standardize data for linear models (not needed for tree-based models)
    scaler = StandardScaler()
    X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

    # 1Ô∏è‚É£ Correlation with Target
    correlation_pearson = X.corrwith(y, method='pearson')
    correlation_spearman = X.corrwith(y, method='spearman')

    # 2Ô∏è‚É£ Random Forest Feature Importance
    rf = RandomForestRegressor(n_estimators=100, random_state=42)
    rf.fit(X, y)
    rf_importance = pd.Series(rf.feature_importances_, index=X.columns)

    # 3Ô∏è‚É£ XGBoost Feature Importance
    xgb = XGBRegressor(n_estimators=100, random_state=42)
    xgb.fit(X, y)
    xgb_importance = pd.Series(xgb.feature_importances_, index=X.columns)

    # 4Ô∏è‚É£ SHAP Values for XGBoost
    explainer = shap.Explainer(xgb)
    shap_values = explainer(X, check_additivity=False)
    shap_importance = pd.Series(np.abs(shap_values.values).mean(axis=0), index=X.columns)

    # 5Ô∏è‚É£ RMSE and R¬≤ for Each Feature (Simple Linear Regression)
    r2_scores = pd.Series([
        LinearRegression().fit(X_scaled[[feature]], y).score(X_scaled[[feature]], y) 
        for feature in X_scaled.columns], index=X_scaled.columns)

    rmse_scores = pd.Series([
            np.sqrt(mean_squared_error(y, LinearRegression().fit(X_scaled[[feature]], y).predict(X_scaled[[feature]]))) 
            for feature in X_scaled.columns], index=X_scaled.columns)

    # 6Ô∏è‚É£ Combine Everything into a Table
    feature_table = pd.DataFrame({
        'Pearson_Corr': correlation_pearson,
        'Spearman_Corr': correlation_spearman,
        'RF_Importance': rf_importance,
        'XGB_Importance': xgb_importance,
        'SHAP_Importance': shap_importance,
        'RMSE': rmse_scores,
        'R2': r2_scores})

    # 7Ô∏è‚É£ Create Status Table to Keep Top Features
    status_table = pd.DataFrame(index=X.columns)
    # Identify top_n features per metric and mark as 'keep'
    for metric in ['Pearson_Corr', 'Spearman_Corr', 'RF_Importance', 'XGB_Importance', 'SHAP_Importance']:
        if 'Corr' in metric:  # Use absolute values for correlation metrics
            top_features = feature_table[metric].abs().nlargest(top_n)
        else:
            top_features = feature_table[metric].nlargest(top_n)
        
        status_table[metric] = status_table.index.to_series().map(
            lambda x: f"keep ({feature_table[metric][x]:.4f})" if x in top_features.index else '')

    # Select top_n based on RMSE (smallest values) and R¬≤ (largest values)
    top_r2 = r2_scores.nlargest(top_n)
    top_rmse = rmse_scores.nsmallest(top_n)

    status_table['R2'] = status_table.index.to_series().map(
        lambda x: f"keep ({top_r2[x]:.4f})" if x in top_r2.index else '')
    status_table['RMSE'] = status_table.index.to_series().map(
        lambda x: f"keep ({top_rmse[x]:.4f})" if x in top_rmse.index else '')

    # Keep only rows with at least one 'keep' entry
    status_table = status_table.loc[
        status_table.apply(lambda row: any(row.str.contains('keep')), axis=1)]

    return status_table
```

#### <span style="background-color:#000027; padding:5px; border-radius:5px;">**5. Lags**</span> <a id='lags'></a>  

In time series, rows are not independent (in fact, we wish to use the dependence to help with our modelling!)
- When working with time series we wish to use information from past values to help make predictions about future values
- For supervised machine learning, we need an input feature matrix, X, and a response vector, y.
- We can create our features (X) by adding columns to our dataframe that are lagged versions of the response

```python
def create_time_series_features(df, lags):
    df = df.copy()
    df.index = pd.to_datetime(df.index)  # Ensure datetime index
    df["time_idx"] = np.arange(len(df))  # Create a simple trend feature

    # Extract basic time-based features
    df["year"] = df.index.year
    df["month"] = df.index.month

    # Create a dictionary to store new features for better performance
    new_features = {}

    for col in df.columns:  
        if col.startswith("P"):  # Only apply to product sales columns
            # Lag features
            for lag in lags:
                new_features[f"{col}_lag_{lag}"] = df[col].shift(lag)
            
            # Moving Average Features
            for window in lags:
                new_features[f"{col}_ma_{window}"] = df[col].rolling(window=window).mean()

    # Concatenate new features to the original DataFrame
    df = pd.concat([df, pd.DataFrame(new_features)], axis=1)
    
    # Drop NaNs from lag-based features (Optional, depending on your use case)
    df = df.dropna()
    
    return df

def calculate_adjusted_r2_and_summary(df, target, lags):
    """ Fit OLS model and return adjusted R¬≤ & model summary. """
    if len(df) < 20:
        return -np.inf, None  # Too few data points

    X = df[[f"{target}_lag_{lag}" for lag in lags]]
    y = df[target]

    X = sm.add_constant(X)  # Add constant for OLS
    try:
        model = sm.OLS(y, X).fit()
        return model.rsquared_adj, model.summary()
    except Exception as e:
        print(f"Error with lags {lags}: {e}")
        return -np.inf, None

def find_best_lags(df, target):
    """ Find the best lag combination maximizing adjusted R¬≤. """
    best_r2 = -float('inf')
    best_lags = None
    best_summary = None
    results = []

    candidate_lags = list(range(1, 12)) + [12, 17, 18]  

    for i in range(1, len(candidate_lags) + 1):
        for lag_combo in combinations(candidate_lags, i):
            if len(lag_combo) > 8: 
                continue

            df_lags = create_time_series_features(df, lags=list(lag_combo))
            adj_r2, summary = calculate_adjusted_r2_and_summary(df_lags, target, list(lag_combo))
            
            results.append({"Lags": lag_combo, "Adjusted R¬≤": adj_r2})
            
            if adj_r2 > best_r2:
                best_r2 = adj_r2
                best_lags = list(lag_combo)
                best_summary = summary

    # Convert results to DataFrame for easy visualization
    results_df = pd.DataFrame(results).sort_values(by="Adjusted R¬≤", ascending=False)

    print(f"Best Lags: {best_lags}, Adjusted R¬≤: {best_r2:.4f}")
    if best_summary:
        print(best_summary)

    return best_lags
```

#### <span style="background-color:#000027; padding:5px; border-radius:5px;">**6. Outliers**</span> <a id='outliers'></a>  

```python
def plot_time_series_with_outliers(dfs, title="Time Series Outliers"):
    num_products = len(dfs)
    fig = make_subplots(rows=1, cols=num_products, subplot_titles=list(dfs.keys()), horizontal_spacing=0.02)

    for i, (key, df) in enumerate(dfs.items()):
        # Original data
        fig.add_trace(go.Scatter(
            y=df[key], mode="lines", name="Original Data"),
            row=1, col=i+1)

        # Winsorization bounds
        lower_bound = df[key].quantile(0.05)
        upper_bound = df[key].quantile(0.95)

        fig.add_trace(go.Scatter(
            y=[lower_bound] * len(df), mode="lines", name="Lower Bound", line=dict(dash="dash", color="red")),
            row=1, col=i+1)

        fig.add_trace(go.Scatter(
            y=[upper_bound] * len(df), mode="lines", name="Upper Bound", line=dict(dash="dash", color="red")),
            row=1, col=i+1)

    fig.update_layout(
        title_text=title,
        height=500,
        width=num_products * 600,  # Adjust the width dynamically
        showlegend=False,
        xaxis=dict(showticklabels=False),
        yaxis_title="Value"
    )
    fig.show()
```

```python
# Step 1: outliers before Winsorization
plot_time_series_with_outliers(product_dfs, title="Outliers Before Winsorization")
```

```python
# Step 2: Apply Winsorization only to the product column
def winsorize_function(df, product_key, lower_quantile=0.05, upper_quantile=0.95):
    if product_key in df.columns:  # Apply only if the product column exists
        q1 = df[product_key].quantile(lower_quantile)
        q3 = df[product_key].quantile(upper_quantile)
        df[product_key] = df[product_key].clip(lower=q1, upper=q3)
        print(f"{product_key}: Winsorized Bounds -> Lower = {q1:.2f}, Upper = {q3:.2f}")
    return df

# Apply Winsorization only on product-named columns
winsorized_product_dfs = {key: winsorize_function(df.copy(), key) for key, df in product_dfs.items()}
```
