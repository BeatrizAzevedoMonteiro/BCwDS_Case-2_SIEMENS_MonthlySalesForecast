```python
df_market.index = df_market.index.astype(str)

# list of countries
countries = ["China", "France", "Germany", "Italy", "Japan", "Switzerland", "UK", "US", "Europe"]

fig = go.Figure()

# add a line for each country
for country, column in zip(countries, shipment_columns):
    fig.add_trace(go.Scatter(x=df_market.index, y=df_market[column], mode='lines', name=country))

# customize layout
fig.update_layout(
    title="Shipment Trends by Country",
    xaxis_title="Year",
    yaxis_title="Total Shipments",
    legend_title="Country",
    template="plotly_white",
    hovermode="x",
    width=1000, height=600
)

fig.show()
```

- Later analyzed individually.

```python
num_countries = len(countries)

# create a subplot layout 
fig = make_subplots(
    rows=1, cols=num_countries, 
    subplot_titles=countries,  
    horizontal_spacing=0.02 
)

# loop through each country and add traces for the shipment trend
for i, (country, column) in enumerate(zip(countries, shipment_columns)):
    # add Shipments trace for each country
    fig.add_trace(go.Scatter(
        x=df_market.index, 
        y=df_market[column], 
        mode='lines', 
        name=f'{country} Shipments',
        line=dict(color="skyblue", width=2)
    ), row=1, col=i+1)

fig.update_layout(
    title='Shipment Trend for Different Countries',
    height=400, 
    width=600 * num_countries, 
    showlegend=True,  
    template="plotly_white", 
    title_x=0.5 
)

fig.show()
```

- There is a clear steep ascending trend in China.
- For the other countries, there is still a clear ascending trend, but not as steep as the one identified in China.
- In both 2010 and 2020, there are noticeable negative peaks. These declines may be attributed to the sovereign debt crisis in 2010 and the COVID-19 pandemic in 2020.
- The UK has a drop in 2020, which might be caused due to the Brexit.
- Switzerland lacks a clear pattern, showing irregular trends compared to the other countries.

- We calculate the monthly growth rate of the variable by computing the percentage change for each month. Next, we calculate the average of these monthly growth rates. Using this average growth rate, we can fill in any missing values by multiplying the value from the prior month by (1 + average growth rate). This method allows us to estimate the missing value based on the observed trend in the data.

- We also exclude the last month (2022-04-01) from this calculation because that is the month for which we are trying to fill in the missing values.

```python
# exclude the last month (2022-04-01) for calculation
df_market_filtered_excluding_last = df_market_filtered[df_market_filtered.index != '2022-04-01']
```

```python
def feature_selection(df, target_col):
    """
    Performs feature selection by calculating various importance metrics.
    """
    # Prepare X and y
    X = df.drop(columns=[target_col])
    y = df[target_col]

    # Standardize data
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # 1️⃣ Correlation with Target
    correlation_pearson = X.corrwith(y, method='pearson')
    correlation_spearman = X.apply(lambda col: spearmanr(col, y)[0])

    # 3️⃣ Random Forest Feature Importance
    rf = RandomForestRegressor(n_estimators=100, random_state=42)
    rf.fit(X, y)
    rf_importance = pd.Series(rf.feature_importances_, index=X.columns)

    # 4️⃣ XGBoost Feature Importance
    xgb = XGBRegressor(n_estimators=100, random_state=42)
    xgb.fit(X, y)
    xgb_importance = pd.Series(xgb.feature_importances_, index=X.columns)

    # 5️⃣ SHAP Values for XGBoost
    explainer = shap.Explainer(xgb)
    shap_values = explainer(X_scaled)
    shap_importance = pd.Series(np.abs(shap_values.values).mean(axis=0), index=X.columns)

    # 6️⃣ RMSE and R² for Each Feature (Simple Linear Regression)
    rmse_scores = {}
    r2_scores = {}

    for feature in X.columns:
        X_feature = X[[feature]]  # Select one feature at a time
        model = LinearRegression()
        model.fit(X_feature, y)
        y_pred = model.predict(X_feature)

        rmse_scores[feature] = np.sqrt(mean_squared_error(y, y_pred))
        r2_scores[feature] = r2_score(y, y_pred)

    rmse_df = pd.Series(rmse_scores)
    r2_df = pd.Series(r2_scores)

    # 7️⃣ Combine Everything into a Table
    feature_table = pd.DataFrame({
        'Pearson_Corr': correlation_pearson,
        'Spearman_Corr': correlation_spearman,
        'RF_Importance': rf_importance,
        'XGB_Importance': xgb_importance,
        'SHAP_Importance': shap_importance,
        'RMSE': rmse_df,
        'R2': r2_df
    }).sort_values(by='Spearman_Corr', ascending=False)

    return feature_table
```
