```python
missing_values = df_market.isnull().sum().to_frame(name='Total')
missing_values['Percentage'] = (missing_values['Total'] / len(df_market) * 100).round(2)
missing_values = missing_values[missing_values['Total'] > 0]
missing_values['Percentage'] = missing_values['Percentage'].astype(str) + '%'
missing_values.sort_values(by='Total', ascending=False, inplace=True)
print(missing_values)
```

```python
# Check for any negative values in the entire dataset
positive_values = (df_market > 0).any().any()

if positive_values:
    print("Positive values exist in the dataset.")
else:
    print("No positive values found in the dataset.")
```

```python
df_sales = pd.read_csv('Data/Case2_Sales data.csv', delimiter=';')
df_sales.head()
```

```python
df_sales.describe()
```

```python
df_sales.shape
```

```python
df_sales.info()
```

```python
missing_values_sales = df_sales.isnull().sum()

features_with_missing_values_sales = missing_values_sales[missing_values_sales > 0]
features_with_missing_values_sales
```

```python
num_unique_dates = df_sales['DATE'].nunique()
print(num_unique_dates)
```

```python
df_sales.duplicated().sum()
```

```python
missing_values_sales = df_sales.isnull().sum().to_frame(name='Total')
missing_values_sales['Percentage'] = (missing_values_sales['Total'] / len(df_sales) * 100).round(2)
missing_values_sales = missing_values_sales[missing_values_sales['Total'] > 0]
missing_values_sales['Percentage'] = missing_values_sales['Percentage'].astype(str) + '%'
missing_values_sales.sort_values(by='Total', ascending=False, inplace=True)
print(missing_values_sales)
```

- There are no missing values in 'df_sales'.

```python
# remove the # symbol
df_sales['Mapped_GCK'] = df_sales['Mapped_GCK'].astype(str).str.replace('#', '', regex=False)
# unique value of Mapped_GCK
df_sales['Mapped_GCK'].value_counts()
```

##### <span style="background-color:#235987; padding:5px; border-radius:5px;">**Correct Sales Data Types**</span> <a id='data-loading-and-description'></a>   

##### Click [here](#table-of-contents) ‚¨ÜÔ∏è to return to the Index.


```python
df_sales['DATE'] = pd.to_datetime(df_sales['DATE'], format='%d.%m.%Y')
# replace commas with dots
df_sales['Sales_EUR'] = df_sales['Sales_EUR'].astype(str).str.replace(',', '.', regex=False).astype(float)
```

```python
df_sales.head()
```

```python
# check for negative values in numeric columns
negative_values = df_sales[
    (df_sales['Sales_EUR'] < 0)
]
print(f"Negative Values Found: {negative_values.shape[0]} records")
```

```python
negative_values
```

```python
negative_values['Mapped_GCK'].value_counts()
```

```python
# extract the year from 'DATE'
negative_by_year = negative_values.groupby(negative_values['DATE'].dt.year)['Mapped_GCK'].count()
print(negative_by_year)
```

```python
# extract the year and month from 'DATE'
negative_values['YearMonth'] = negative_values['DATE'].dt.to_period('M')
# group them and count the negative values in 'Mapped_GCK'
negative_by_month = negative_values.groupby('YearMonth')['Mapped_GCK'].count().reset_index()
print(negative_by_month)
```

#### <span style="background-color:#235987; padding:5px; border-radius:5px;">**Sales Data Visualization**</span> <a id='data-loading-and-description'></a>   

##### Click [here](#table-of-contents) ‚¨ÜÔ∏è to return to the Index.


##### <span style="background-color:#235987; padding:5px; border-radius:5px;">**Negative Sales**</span> <a id='data-loading-and-description'></a>  

```python
x_labels = negative_by_month['YearMonth']
x_ticks = np.arange(0, len(x_labels), 6)  # every 6th month

# plotting the results
plt.figure(figsize=(12, 4))
plt.plot(x_labels, negative_by_month['Mapped_GCK'], marker='o', color='cornflowerblue', linestyle='-', linewidth=2)

# formatting the plot
plt.title('Negative Mapped_GCK Counts by Month')
plt.xlabel('Year-Month')
plt.ylabel('Negative Mapped_GCK Count')
plt.xticks(ticks=x_ticks, labels=x_labels[x_ticks], rotation=0)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
negative_values['DATE'] = pd.to_datetime(negative_values['DATE'])

# extract the year from 'DATE'
negative_values['Year'] = negative_values['DATE'].dt.year

# group by 'Year' and count the occurrences of negative values in 'Mapped_GCK'
negative_by_year = negative_values.groupby('Year')['Mapped_GCK'].count().reset_index()

# plotting the results
plt.figure(figsize=(8, 4))
plt.bar(negative_by_year['Year'].astype(str), negative_by_year['Mapped_GCK'], color='cornflowerblue')
plt.title('Negative Mapped_GCK Counts by Year')
plt.xlabel('Year')
plt.ylabel('Negative Mapped_GCK Count')
plt.xticks(rotation=0) 
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

##### <span style="background-color:#235987; padding:5px; border-radius:5px;">**All Sales**</span> <a id='data-loading-and-description'></a>  

```python
# group the data by 'YearMonth' and sum the total sales
df_sales['YearMonth'] = df_sales['DATE'].dt.to_period('M')
sales_by_month = df_sales.groupby('YearMonth')['Sales_EUR'].sum().reset_index()
# convert 'YearMonth' to string for x-axis labels
sales_by_month['YearMonth'] = sales_by_month['YearMonth'].astype(str)
x_labels = sales_by_month['YearMonth']
x_ticks = np.arange(0, len(x_labels), 6)  # every 6th month

# plotting the results
plt.figure(figsize=(12, 4))
plt.plot(x_labels, sales_by_month['Sales_EUR'], marker='o', color='cornflowerblue', linestyle='-', linewidth=2)
plt.title('Total Sales (EUR) by Month')
plt.xlabel('Year-Month')
plt.ylabel('Total Sales (EUR)')
plt.xticks(ticks=x_ticks, labels=x_labels[x_ticks], rotation=0)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# extract the year from 'DATE'
df_sales['Year'] = df_sales['DATE'].dt.year

# group by 'Year' and sum the 'Sales_EUR'
sales_by_year = df_sales.groupby('Year')['Sales_EUR'].sum().reset_index()

# plotting the results
plt.figure(figsize=(8, 4))
plt.bar(sales_by_year['Year'].astype(str), sales_by_year['Sales_EUR'], color='cornflowerblue')
plt.title('Total Sales (EUR) by Year')
plt.xlabel('Year')
plt.ylabel('Total Sales (EUR)')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
